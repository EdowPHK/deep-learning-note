## About `Automatic Differentiation`

### Backward for Scalar

```python
import torch
```

```python
x = torch.arange(4.0, requires_grad=True)
```

**Using `.grad` to access a tensor's gradient.**

```python
x.grad
```

```python
y = 2 * torch.dot(x, x)
```

**Using `.backward()` to take the gradient: $\nabla_{\mathbf{X}} f(\mathbf{x})$**

```python
y.backward()
x.grad
```

**Using `.grad.zero_()` to initialize the gradient.**

```python
x.grad.zero_()
```

### Backward for Non-Scalar Variables

**Since the following `y` is a vector, we should set a specified vector called `gradient` to let `backward()` compute $\mathbf{v}^\top \partial_{\mathbf{x}} \mathbf{y}$ rather than $\partial_{\mathbf{x}} \mathbf{y}$, which is equivalent to multiplying `y` by $\mathbf{v}^\top$, then produce the `backward()`.**

```python
y = x * x
y.backward(gradient=torch.ones(len(y)))
x.grad
```

### Detaching Computation

**Using `.detach()` to wipe out the provenance(how it was created) to see it as a constant.**

```python
x.grad.zero_()
y = x * x
u = y.detach()
z = u * x

z.backward(gradient=torch.ones(len(z)))
x.grad == u
```

**Which should be noticed that the computational graph leading to `y` presists and thus we can calculate the gradient of `y` with respect to `x`.**

```python
x.grad.zero_()
y.backward(gradient=torch.ones(len(y)))
x.grad == 2 * x
```
