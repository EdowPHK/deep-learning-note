### Reading the Dataset

**Using `pd.read_csv(file_path)` to load the dataset.**

```python
data = pd.read_csv('/home/yanzi888/d2l/d2l-en/data/house_tiny.csv')
print(data)
```

### Data Preparation

**In Supervised Learning, our first step is to separate columns into inputs and outputs with *Slicing(切片)*, Using `data.iloc[]` to access these multiple categories of data.**

**For missing categorical values, using `pd.get_dummies(inputs, dummy_na=True)` treat `NaN` as a category through setting extra columns.**

```python
inputs, targets = data.iloc[:, 0:2], data.iloc[:, 2]
inputs = pd.get_dummies(inputs, dummy_na=True)
inputs_false = pd.get_dummies(inputs, dummy_na=False)
print(inputs),print(inputs_false)
```

**For missing numerical values, using `.fillna(.mean())` to replace `NaN` entries with the mean value of the corresponding column.**

```python
inputs = inputs.fillna(inputs.mean())
print(inputs)
```
