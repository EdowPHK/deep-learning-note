# Linear Algebra

```python
import torch
```

**Using `.sum(axis=n, keepdims=True)` to sum up elements of specified dimension with the number of axes unchanged.**

```python
A = torch.arange(24, dtype=torch.float32).reshape(2,3,4)
A
```

```python
A.sum(axis=0), A.sum(axis=1), A.sum(axis=2)
```

```python
A.sum(axis=0, keepdims=True), A.sum(axis=1, keepdims=True), A.sum(axis=2, keepdims=True)
```
