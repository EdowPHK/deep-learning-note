# **Preliminaries**

## **-About `ndarray`**

### Getting started

```python
import torch
```

**USING `torch.arange` to create a row tensor: *[0,1,2,···,10]***

```python
x = torch.arange(10)
x
```

**USING `.shape` to access a tensor's shape**

```python
x.shape
```

**USING `.numel` to get the number of elements in tensor**

```python
x.numel()
```

**USING `.reshape()` to change tensors shape specifically**

```python
x.reshape(2, -1)
```

**USING `torch.zeros` / `torch.ones` to generate tensors with all 0s / 1s**

```python
a = torch.zeros((2, 3, 4))
```

**USING `torch.zeros_like(a)`/`torch.ones_like(a)` to generate a tensor having the same shape as a.**

```python
torch.ones_like(a)
```

**USING `torch.randn()` to generate tensors with elements drawn from a standard Gaussian (normal) distribution with mean 0 and standard deviation 1**

```python
torch.randn(3)
```

**USING `torch.tensor` to generate tensors precisely as specified**

```python
torch.tensor([[1,2,3],[2,4,5]])
```

### Indexing and Slicing

```python
x = torch.tensor([[1,2,3],[2,3,4],[3,4,5],[4,5,6]])
```

**USING `x[a:b]` to acess the specified elements in tensor x, by which can get the *a_th* element to *(b-1)_th* element**

```python
x[0:2]
```

**NOTICE - Different output between `x[-2]` and `x[-2:-1]`**

```python
x[-2]
```

```python
x[-2:-1]
```

**We could change the elements precisely as well :**

```python
x[2,0] = 6
x
```

```python
x[:2, :] = 12
x
```

### Operations

**`torch.exp()` / `+` / `-` / `*` / `/` / `**`**

```python
x = torch.tensor([1,2,4,8])
y = torch.tensor([2,2,2,2])
torch.exp(y), x + y, x - y, x * y, x / y, x ** y
```

**USING `torch.cat((x,y), dim=)` to concatenate two tensors.(*shape must be same except for the concatenating dimension*)**

```python
X = torch.arange(12, dtype=torch.float32).reshape(3,4)
Y = torch.tensor([[2.0, 1, 3, 4, 5], [1, 2, 3, 4, 5], [4, 3, 2, 1, 5]])
torch.cat((X, Y),dim=1)
```

```python
X = torch.arange(12, dtype=torch.float32).reshape(3,4)
Y = torch.tensor([[2.0, 1, 3, 4], [1, 2, 3, 4], [4, 3, 2, 1]])
torch.cat((X, Y),dim=0)
```

**USING `==` to judge if two tensors are equal.**

```python
X == Y
```

**USING `.sum()` to summing ALL the elements in the tensor and generate a tensor with only ONE element.**

```python
X.sum()
```

### Broadcasting

**With `Broadcasting mechanism` to perform elementwise binary operations on tensors shapes differ.**

```python
a = torch.arange(3).reshape((3,1))
b = torch.arange(2).reshape((1,2))
a, b, a + b
```

### Memory

**USING `id()` to access tensor's location in memory,and USING `Z[:] = X + Y` to performing in-place operations.**

```python
Z = torch.tensor([1, 2, 3])
print('id(Z):',id(Z))
Z[:] = Z + 1
print('id(Z):',id(Z))
```

### Conversion to Other Python Objects

**USING `torch.from_numpy(a)` to change the type of `a` from ndarray to tensor.**

```python
A = X.numpy()
B = torch.from_numpy(A)
type(A), type(B)
```
