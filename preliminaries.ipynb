# **Preliminaries**

## **-About `ndarray`**

```python
import torch
```

**USING `torch.arange` to create a row tensor: *[0,1,2,···,10]***

```python
x = torch.arange(10)
x
```

**USING `.shape` to visit a tensor's shape**

```python
x.shape
```

**USING `.numel` to get the number of elements in tensor**

```python
x.numel()
```

**USING `.reshape()` to change tensors shape specifically**

```python
x.reshape(2, -1)
```

**USING `torch.zeros` / `torch.ones` to create tensors with all 0s / 1s**

```python
torch.zeros((2, 3, 4))
```

```python
torch.ones((2, 3, 4))
```
