# **Preliminaries**

## **-About `ndarray`**

### Getting started

```python
import torch
```

**USING `torch.arange` to create a row tensor: *[0,1,2,···,10]***

```python
x = torch.arange(10)
x
```

**USING `.shape` to access a tensor's shape**

```python
x.shape
```

**USING `.numel` to get the number of elements in tensor**

```python
x.numel()
```

**USING `.reshape()` to change tensors shape specifically**

```python
x.reshape(2, -1)
```

**USING `torch.zeros` / `torch.ones` to generate tensors with all 0s / 1s**

```python
torch.zeros((2, 3, 4))
```

```python
torch.ones((2, 3, 4))
```

**USING `torch.randn()` to generate tensors with elements drawn from a standard Gaussian (normal) distribution with mean 0 and standard deviation 1**

```python
torch.randn(3)
```

**USING `torch.tensor` to generate tensors precisely as specified**

```python
torch.tensor([[1,2,3],[2,4,5]])
```

### Indexing and Slicing

```python
x = torch.tensor([[1,2,3],[2,3,4],[3,4,5],[4,5,6]])
```

**USING `x[a:b]` to acess the specified elements in tensor x, by which can get the *a_th* element to *(b-1)_th* element**

```python
x[0:2]
```

**NOTICE - Different output between `x[-2]` and `x[-2:-1]`**

```python
x[-2]
```

```python
x[-2:-1]
```

**We could change the elements precisely as well :**

```python
x[2,0] = 6
x
```
